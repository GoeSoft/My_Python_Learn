```python
from transformers import MarianMTModel, MarianTokenizer

# Три английских фрагмента
english_texts = [
    "Artificial intelligence is transforming the world.",
    "Machine learning models require large amounts of data.",
    "Natural language processing enables computers to understand human speech."
]

# Языковые пары: (код языка, название модели, название языка)
target_languages = [
    ("de", "немецкий"),
    ("fr", "французский"),
    ("ru", "русский")
]

# Для каждого целевого языка
for lang_code, lang_name in target_languages:
    print(f"\n🌍 Перевод на {lang_name} ({lang_code}):")
    print("-" * 50)

    model_name = f"Helsinki-NLP/opus-mt-en-{lang_code}"
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)

    for i, text in enumerate(english_texts, 1):
        # Токенизация и перевод
        inputs = tokenizer(text, return_tensors="pt", padding=True)
        translated = model.generate(**inputs)
        result = tokenizer.decode(translated[0], skip_special_tokens=True)
        print(f"{i}. {result}")
```

```python
from transformers import AutoTokenizer

# Загружаем токенизатор Saiga (на основе LLaMA)
model_name = "IlyaGusev/saiga2_7b"
tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=False)

# Примеры предложений
sentences = [
    "Привет! Как дела?",
    "Машинное обучение — это здорово.",
    "Hello! How are you?",
    "Artificial intelligence changes everything."
]

print("🔍 Токенизация и детокенизация с Saiga:")
print("=" * 60)

for sent in sentences:
    tokens = tokenizer.encode(sent)  # Получаем ID токенов
    decoded = tokenizer.decode(tokens)  # Восстанавливаем текст

    print(f"🔤 Исходный текст: {sent}")
    print(f"🔢 Токены (ID): {tokens}")
    print(f"🔤 Детокенизация: {decoded}")
    print("-" * 60)
```
