```python
from gtts import gTTS
import os

# –¢–µ–∫—Å—Ç –Ω–∞ —Ä—É—Å—Å–∫–æ–º (–∫–∏—Ä–∏–ª–ª–∏—Ü–µ–π)
text = "–ú–Ω–æ–≥–∏–µ –∑–Ω–∞—é—Ç –º–Ω–æ–≥–æ, –Ω–æ –Ω–∏–∫—Ç–æ –Ω–µ –∑–Ω–∞–µ—Ç –≤—Å—ë"

# –Ø–∑—ã–∫–∏ –¥–ª—è —Å–∏–Ω—Ç–µ–∑–∞
languages = {
    "ru": "–†—É—Å—Å–∫–∏–π (–Ω–æ—Å–∏—Ç–µ–ª—å)",
    "en": "–ê–Ω–≥–ª–∏–π—Å–∫–∏–π (–∏–º–∏—Ç–∞—Ü–∏—è –∞–∫—Ü–µ–Ω—Ç–∞)",
    "fr": "–§—Ä–∞–Ω—Ü—É–∑—Å–∫–∏–π (–∏–º–∏—Ç–∞—Ü–∏—è –∞–∫—Ü–µ–Ω—Ç–∞)"
}

audio_files = {}

for lang_code, lang_name in languages.items():
    print(f"–°–∏–Ω—Ç–µ–∑ —Ä–µ—á–∏ –Ω–∞ —è–∑—ã–∫–µ: {lang_name} ({lang_code})")
    try:
        tts = gTTS(text=text, lang=lang_code, slow=False)
        filename = f"speech_{lang_code}.mp3"
        tts.save(filename)
        audio_files[lang_code] = filename
        print(f"‚úÖ –°–æ—Ö—Ä–∞–Ω–µ–Ω–æ: {filename}")
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –¥–ª—è {lang_code}: {e}")

print("\n–ì–æ—Ç–æ–≤–æ! –ê—É–¥–∏–æ—Ñ–∞–π–ª—ã —Å–æ–∑–¥–∞–Ω—ã.")
```

```python
from IPython.display import Audio, display

print("üéß –ü—Ä–æ—Å–ª—É—à–∞–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã:\n")

for lang_code, filename in audio_files.items():
    lang_name = languages[lang_code]
    print(f"\n--- {lang_name} ---")
    display(Audio(filename, autoplay=False))

print("\n" + "="*70)
print("üìù –í–´–í–û–î:")
print("="*70)
print("""
1. –†—É—Å—Å–∫–∞—è –º–æ–¥–µ–ª—å (ru) –ø—Ä–æ–∏–∑–Ω–æ—Å–∏—Ç —Ç–µ–∫—Å—Ç –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ ‚Äî —ç—Ç–æ —ç—Ç–∞–ª–æ–Ω.

2. –ê–Ω–≥–ª–∏–π—Å–∫–∞—è –º–æ–¥–µ–ª—å (en) –ø—ã—Ç–∞–µ—Ç—Å—è –ø—Ä–æ—á–∏—Ç–∞—Ç—å –∫–∏—Ä–∏–ª–ª–∏—Ü—É –∫–∞–∫ –ª–∞—Ç–∏–Ω–∏—Ü—É:
   - –ë—É–∫–≤–∞ ¬´–ú¬ª ‚Üí –∑–≤—É—á–∏—Ç –∫–∞–∫ /m/
   - –ù–æ ¬´–∂¬ª, ¬´—ã¬ª, ¬´—å¬ª ‚Äî –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç –≤ –∞–Ω–≥–ª–∏–π—Å–∫–æ–º ‚Üí –∑–∞–º–µ–Ω—è—é—Ç—Å—è –Ω–∞ –±–ª–∏–∂–∞–π—à–∏–µ –∑–≤—É–∫–∏ –∏–ª–∏ –ø—Ä–æ–ø—É—Å–∫–∞—é—Ç—Å—è.
   - –ü–æ–ª—É—á–∞–µ—Ç—Å—è ¬´Mnogee znayet mnogo...¬ª —Å –∏—Å–∫–∞–∂–µ–Ω–∏—è–º–∏ ‚Üí –ø–æ—Ö–æ–∂–µ –Ω–∞ **—Å–∏–ª—å–Ω—ã–π –∞–∫—Ü–µ–Ω—Ç**.

3. –§—Ä–∞–Ω—Ü—É–∑—Å–∫–∞—è –º–æ–¥–µ–ª—å (fr) –≤–µ–¥—ë—Ç —Å–µ–±—è –∞–Ω–∞–ª–æ–≥–∏—á–Ω–æ:
   - –ú—è–≥–∫–∏–µ —Å–æ–≥–ª–∞—Å–Ω—ã–µ, –Ω–æ—Å–æ–≤—ã–µ –≥–ª–∞—Å–Ω—ã–µ ‚Üí –∏—Å–∫–∞–∂–∞—é—Ç —Ä—É—Å—Å–∫–∏–µ –∑–≤—É–∫–∏.
   - –ù–∞–ø—Ä–∏–º–µ—Ä, ¬´–∑–Ω–∞—é—Ç¬ª –º–æ–∂–µ—Ç –ø—Ä–æ–∑–≤—É—á–∞—Ç—å –∫–∞–∫ ¬´–∑–Ω—ë¬ª —Å –Ω–æ—Å–æ–≤—ã–º –æ—Ç—Ç–µ–Ω–∫–æ–º.

4. –û–¥–Ω–∞–∫–æ —ç—Ç–æ **–Ω–µ –Ω–∞—Å—Ç–æ—è—â–∏–π –∞–∫—Ü–µ–Ω—Ç**, –∞ **–∞—Ä—Ç–µ—Ñ–∞–∫—Ç –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø–æ–∑–Ω–∞–≤–∞–Ω–∏—è –∞–ª—Ñ–∞–≤–∏—Ç–∞**.
   - –ù–æ—Å–∏—Ç–µ–ª—å –∞–Ω–≥–ª–∏–π—Å–∫–æ–≥–æ, –≥–æ–≤–æ—Ä—è—â–∏–π –ø–æ-—Ä—É—Å—Å–∫–∏, –≤—Å—ë —Ä–∞–≤–Ω–æ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç **—Ä—É—Å—Å–∫—É—é —Ñ–æ–Ω–µ—Ç–∏–∫—É**, –ø—Ä–æ—Å—Ç–æ —Å –æ—à–∏–±–∫–∞–º–∏.
   - –ê `gTTS` –∑–¥–µ—Å—å –ø—Ä–æ—Å—Ç–æ ¬´–ª–æ–º–∞–µ—Ç¬ª —Å–ª–æ–≤–∞, –ø–æ—Ç–æ–º—É —á—Ç–æ –Ω–µ –∑–Ω–∞–µ—Ç –∫–∏—Ä–∏–ª–ª–∏—Ü—É –≤ –∫–æ–Ω—Ç–µ–∫—Å—Ç–µ –∞–Ω–≥–ª–∏–π—Å–∫–æ–≥–æ.

5. **–í—ã–≤–æ–¥**: —Ç–∞–∫–∏–º —Å–ø–æ—Å–æ–±–æ–º –º–æ–∂–Ω–æ —Å–æ–∑–¥–∞—Ç—å **–∫–∞—Ä–∏–∫–∞—Ç—É—Ä–Ω–æ–µ –ø–æ–¥–æ–±–∏–µ –∞–∫—Ü–µ–Ω—Ç–∞**, –Ω–æ –Ω–µ —Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω–æ–µ –ø—Ä–æ–∏–∑–Ω–æ—à–µ–Ω–∏–µ.
   –î–ª—è –∏–º–∏—Ç–∞—Ü–∏–∏ –Ω–∞—Å—Ç–æ—è—â–µ–≥–æ –∞–∫—Ü–µ–Ω—Ç–∞ –Ω—É–∂–Ω—ã:
   - –ú–æ–¥–µ–ª–∏ TTS —Å –ø–æ–¥–¥–µ—Ä–∂–∫–æ–π –∫—Ä–æ—Å—Å-—è–∑—ã–∫–æ–≤–æ–≥–æ –ø—Ä–æ–∏–∑–Ω–æ—à–µ–Ω–∏—è (–Ω–∞–ø—Ä–∏–º–µ—Ä, **YourTTS**, **VITS**),
   - –ò–ª–∏ fine-tuning –Ω–∞ –¥–∞—Ç–∞—Ö ¬´–∏–Ω–æ—Å—Ç—Ä–∞–Ω–µ—Ü –≥–æ–≤–æ—Ä–∏—Ç –ø–æ-—Ä—É—Å—Å–∫–∏¬ª.

–ò—Ç–æ–≥: gTTS **–Ω–µ –ø–æ–¥—Ö–æ–¥–∏—Ç** –¥–ª—è —Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω–æ–π –∏–º–∏—Ç–∞—Ü–∏–∏ –∞–∫—Ü–µ–Ω—Ç–∞, –Ω–æ –º–æ–∂–µ—Ç –±—ã—Ç—å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω –¥–ª—è —é–º–æ—Ä–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö –∏–ª–∏ —É—á–µ–±–Ω—ã—Ö –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–π.
""")
```
