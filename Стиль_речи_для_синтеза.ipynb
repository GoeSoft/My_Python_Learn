```python
# –û–ø–∏—Å–∞–Ω–∏–µ –∂–µ–ª–∞–µ–º–æ–≥–æ —Å—Ç–∏–ª—è —Ä–µ—á–∏
desired_style = """
–†–æ–º–∞–Ω—Ç–∏—á–µ—Å–∫–∏–π —Å—Ç–∏–ª—å —Ä–µ—á–∏ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏–∑—É–µ—Ç—Å—è:
‚Äî –≠–º–æ—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–π –Ω–∞—Å—ã—â–µ–Ω–Ω–æ—Å—Ç—å—é –∏ –ª–∏—Ä–∏–∑–º–æ–º,
‚Äî –ú—è–≥–∫–æ–π, –ø–ª–∞–≤–Ω–æ–π –∏–Ω—Ç–æ–Ω–∞—Ü–∏–µ–π,
‚Äî –í—ã—Ä–∞–∑–∏—Ç–µ–ª—å–Ω—ã–º–∏ –ø–∞—É–∑–∞–º–∏ –∏ –∑–∞–º–µ–¥–ª–µ–Ω–Ω—ã–º —Ç–µ–º–ø–æ–º –≤ –∫–ª—é—á–µ–≤—ã—Ö –º–æ–º–µ–Ω—Ç–∞—Ö,
‚Äî –ê–∫—Ü–µ–Ω—Ç–æ–º –Ω–∞ –æ–±—Ä–∞–∑–∞—Ö, —á—É–≤—Å—Ç–≤–∞—Ö –∏ –≤–Ω—É—Ç—Ä–µ–Ω–Ω–µ–º –ø–µ—Ä–µ–∂–∏–≤–∞–Ω–∏–∏.

–¶–µ–ª—å: –ø–µ—Ä–µ–¥–∞—Ç—å —ç—Ç–æ –Ω–∞—Å—Ç—Ä–æ–µ–Ω–∏–µ —á–µ—Ä–µ–∑ —Å–∏–Ω—Ç–µ–∑–∏—Ä–æ–≤–∞–Ω–Ω—É—é —Ä–µ—á—å.
"""

# –û–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ –≤—ã–±–æ—Ä–∞ —Ç–µ–∫—Å—Ç–∞
text_choice_justification = """
–°—Ç–∏—Ö–æ—Ç–≤–æ—Ä–µ–Ω–∏–µ –ê.–°. –ü—É—à–∫–∏–Ω–∞ ¬´–Ø –ø–æ–º–Ω—é —á—É–¥–Ω–æ–µ –º–≥–Ω–æ–≤–µ–Ω—å–µ‚Ä¶¬ª –≤—ã–±—Ä–∞–Ω–æ –ø–æ—Ç–æ–º—É, —á—Ç–æ:
‚Äî –≠—Ç–æ –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–∏–π –æ–±—Ä–∞–∑–µ—Ü —Ä—É—Å—Å–∫–æ–π —Ä–æ–º–∞–Ω—Ç–∏—á–µ—Å–∫–æ–π –ª–∏—Ä–∏–∫–∏,
‚Äî –¢–µ–∫—Å—Ç –Ω–∞–ø–æ–ª–Ω–µ–Ω —ç–º–æ—Ü–∏—è–º–∏ –≤–æ—Å—Ö–∏—â–µ–Ω–∏—è, –Ω–µ–∂–Ω–æ—Å—Ç–∏ –∏ –≤–æ–∑–≤—ã—à–µ–Ω–Ω–æ–π –ª—é–±–≤–∏,
‚Äî –Ø–∑—ã–∫ –ø–æ—ç—Ç–∏—á–µ–Ω, –æ–±—Ä–∞–∑–µ–Ω –∏ —Ä–∏—Ç–º–∏—á–µ–Ω ‚Äî –∏–¥–µ–∞–ª—å–Ω–æ –æ—Ç—Ä–∞–∂–∞–µ—Ç —Å—É—Ç—å —Ä–æ–º–∞–Ω—Ç–∏—á–µ—Å–∫–æ–≥–æ —Å—Ç–∏–ª—è.
"""

print(" –ñ–µ–ª–∞–µ–º—ã–π —Å—Ç–∏–ª—å —Ä–µ—á–∏:")
print(desired_style)
print("\nüìú –û–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ –≤—ã–±–æ—Ä–∞ —Ç–µ–∫—Å—Ç–∞:")
print(text_choice_justification)
```

```python
from gtts import gTTS

# –°–∞–º —Ç–µ–∫—Å—Ç —Å—Ç–∏—Ö–æ—Ç–≤–æ—Ä–µ–Ω–∏—è
text = """
–Ø –ø–æ–º–Ω—é —á—É–¥–Ω–æ–µ –º–≥–Ω–æ–≤–µ–Ω—å–µ:
–ü–µ—Ä–µ–¥–æ –º–Ω–æ–π —è–≤–∏–ª–∞—Å—å —Ç—ã,
–ö–∞–∫ –º–∏–º–æ–ª—ë—Ç–Ω–æ–µ –≤–∏–¥–µ–Ω—å–µ,
–ö–∞–∫ –≥–µ–Ω–∏–π —á–∏—Å—Ç–æ–π –∫—Ä–∞—Å–æ—Ç—ã.
"""

# –°–∏–Ω—Ç–µ–∑ –Ω–∞ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ
tts = gTTS(text=text, lang='ru', slow=False)
tts.save("romantic_poem.mp3")

print(" –ê—É–¥–∏–æ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–æ –∫–∞–∫ romantic_poem.mp3")
```
