## –§–Ω–∫—Ü–∏–∏ –∏ –Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –Ω–µ–π—Ä–æ—Å–æ—Ç—Ä—É–¥–Ω–∏–∫–∞ —Å –¥–∏–∞–ª–æ–≥–æ–º

```python
def collect_about_me():
    questions = [
        "–ö–∞–∫ –≤–∞—Å –∑–æ–≤—É—Ç? –ù–∞–ø–∏—à–∏—Ç–µ, –ø–æ–∂–∞–ª—É–π—Å—Ç–∞, –≤–∞—à–µ –ø–æ–ª–Ω–æ–µ –§–ò–û.",
        "–ü–æ–¥—Å–∫–∞–∂–∏—Ç–µ –≤–∞—à –Ω–æ–º–µ—Ä —Ç–µ–ª–µ—Ñ–æ–Ω–∞ –¥–ª—è —Å–≤—è–∑–∏.",
        "–í –∫–∞–∫–æ–º –≥–æ—Ä–æ–¥–µ –≤—ã –ø—Ä–æ–∂–∏–≤–∞–µ—Ç–µ?",
        "–£–∫–∞–∂–∏—Ç–µ, –ø–æ–∂–∞–ª—É–π—Å—Ç–∞, –≤–∞—à—É —ç–ª–µ–∫—Ç—Ä–æ–Ω–Ω—É—é –ø–æ—á—Ç—É.",
        "–ö –∫–∞–∫–æ–º—É —É—Ä–æ–≤–Ω—é –ø—Ä–æ–≥—Ä–∞–º–º–∏—Å—Ç–∞ –≤—ã —Å–µ–±—è –æ—Ç–Ω–æ—Å–∏—Ç–µ: Junior, Middle –∏–ª–∏ Senior?",
        "–í –∫–∞–∫–æ–π –æ–±–ª–∞—Å—Ç–∏ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—è –≤—ã —è–≤–ª—è–µ—Ç–µ—Å—å —Å–ø–µ—Ü–∏–∞–ª–∏—Å—Ç–æ–º? –ù–∞–ø—Ä–∏–º–µ—Ä: –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–π –∏–Ω—Ç–µ–ª–ª–µ–∫—Ç, frontend-—Ä–∞–∑—Ä–∞–±–æ—Ç—á–∏–∫, Java-—Ä–∞–∑—Ä–∞–±–æ—Ç—á–∏–∫, QA-—Ç–µ—Å—Ç–∏—Ä–æ–≤—â–∏–∫ –∏ —Ç.–¥.",
        "–í –∫–∞–∫–æ–π –æ–±–ª–∞—Å—Ç–∏ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—è –≤—ã —Å–µ–π—á–∞—Å —Å–æ–≤–µ—Ä—à–µ–Ω—Å—Ç–≤—É–µ—Ç–µ—Å—å –∏–ª–∏ –∏–∑—É—á–∞–µ—Ç–µ —á—Ç–æ-—Ç–æ –Ω–æ–≤–æ–µ?",
        "–†–∞—Å—Å–∫–∞–∂–∏—Ç–µ, –ø–æ–∂–∞–ª—É–π—Å—Ç–∞, –æ –≤–∞—à–µ–º —Ö–æ–±–±–∏. –≠—Ç–æ –ø–æ–º–æ–∂–µ—Ç —Å–¥–µ–ª–∞—Ç—å —Ä–µ–∑—é–º–µ –±–æ–ª–µ–µ –∂–∏–≤—ã–º."
    ]

    answers = {}
    print("üëã –ü—Ä–∏–≤–µ—Ç! –Ø ‚Äî –ù–µ–π—Ä–æ-HR. –ü–æ–º–æ–≥—É —Å–æ—Å—Ç–∞–≤–∏—Ç—å –ø—Ä–æ—Ñ–µ—Å—Å–∏–æ–Ω–∞–ª—å–Ω–æ–µ —Ä–µ–∑—é–º–µ.\n")
    print("–ù–∞—á–Ω—ë–º —Å –±–ª–æ–∫–∞ ¬´–û —Å–µ–±–µ¬ª. –û—Ç–≤–µ—Ç—å—Ç–µ, –ø–æ–∂–∞–ª—É–π—Å—Ç–∞, –Ω–∞ –Ω–µ—Å–∫–æ–ª—å–∫–æ –≤–æ–ø—Ä–æ—Å–æ–≤:\n")

    for i, q in enumerate(questions):
        print(f"HR: {q}")
        while True:
            user_input = input("–í—ã: ").strip()
            if user_input:
                answers[f"q{i+1}"] = user_input
                print("–°–ø–∞—Å–∏–±–æ! –û—Ç–ª–∏—á–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è.\n")
                break
            else:
                print("–ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –≤–≤–µ–¥–∏—Ç–µ –æ—Ç–≤–µ—Ç (—Ö–æ—Ç—è –±—ã –∫—Ä–∞—Ç–∫–∏–π).")

    print("‚úÖ –ë–ª–æ–∫ ¬´–û —Å–µ–±–µ¬ª –ø–æ–ª–Ω–æ—Å—Ç—å—é —Å–æ–±—Ä–∞–Ω. –°–ø–∞—Å–∏–±–æ!")
    return answers
```

# —Å–æ–∑–¥–∞–Ω–∏–µ —Ç–µ–∫—Å—Ç–æ–≤–æ–π —á–∞—Å—Ç–∏ –≤ —Ä–µ–∑—é–º–µ "–¶–µ–ª–∏"

```python
# –ò–Ω—Å—Ç—Ä—É–∫—Ü–∏—è –¥–ª—è GPT, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –ø–æ–¥–∞–≤–∞—Ç—å—Å—è –≤ system –¥–ª—è –±–ª–æ–∫–∞ "–¶–µ–ª–∏"
def collect_goals():
    questions = [
        "–ö–∞–∫—É—é –¥–æ–ª–∂–Ω–æ—Å—Ç—å –≤—ã –∏—â–µ—Ç–µ? –ù–∞–ø—Ä–∏–º–µ—Ä: Middle Python-—Ä–∞–∑—Ä–∞–±–æ—Ç—á–∏–∫, –∞–Ω–∞–ª–∏—Ç–∏–∫ –¥–∞–Ω–Ω—ã—Ö, QA-–∏–Ω–∂–µ–Ω–µ—Ä –∏ —Ç.–¥.",
        "–í –∫–∞–∫–æ–π —Å—Ñ–µ—Ä–µ –∏–ª–∏ –æ—Ç—Ä–∞—Å–ª–∏ –≤—ã —Ö–æ—Ç–µ–ª–∏ –±—ã —Ä–∞–±–æ—Ç–∞—Ç—å? (IT, —Ñ–∏–Ω–∞–Ω—Å—ã, e-commerce, –∑–¥—Ä–∞–≤–æ–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –∏ —Ç.–¥.)",
        "–ß—Ç–æ –¥–ª—è –≤–∞—Å –≤–∞–∂–Ω–æ –≤ —Ä–∞–±–æ—Ç–µ? –ù–∞–ø—Ä–∏–º–µ—Ä: —É–¥–∞–ª—ë–Ω–∫–∞, –∫–∞—Ä—å–µ—Ä–Ω—ã–π —Ä–æ—Å—Ç, –∏–Ω—Ç–µ—Ä–µ—Å–Ω—ã–µ –∑–∞–¥–∞—á–∏, —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å, –∫–æ–º–∞–Ω–¥–∞.",
        "–ï—Å—Ç—å –ª–∏ —É –≤–∞—Å –ø—Ä–µ–¥–ø–æ—á—Ç–µ–Ω–∏—è –ø–æ —Ç–∏–ø—É –∫–æ–º–ø–∞–Ω–∏–∏? (—Å—Ç–∞—Ä—Ç–∞–ø, –∫—Ä—É–ø–Ω–∞—è –∫–æ—Ä–ø–æ—Ä–∞—Ü–∏—è, –ø—Ä–æ–¥—É–∫—Ç, –∞—É—Ç—Å–æ—Ä—Å –∏ —Ç.–¥.)"
    ]

    answers = {}
    print("\n–¢–µ–ø–µ—Ä—å —Å–æ–±–µ—Ä—ë–º –±–ª–æ–∫ ¬´–¶–µ–ª–∏¬ª ‚Äî –æ–Ω –æ—á–µ–Ω—å –≤–∞–∂–µ–Ω –¥–ª—è HR.\n")

    for i, q in enumerate(questions):
        print(f"HR: {q}")
        while True:
            user_input = input("–í—ã: ").strip()
            if user_input:
                answers[f"g{i+1}"] = user_input
                print("–û—Ç–ª–∏—á–Ω–æ! –≠—Ç–æ –ø–æ–º–æ–∂–µ—Ç —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å —Å–∏–ª—å–Ω—É—é —Ü–µ–ª—å.\n")
                break
            else:
                print("–ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –¥–∞–π—Ç–µ —Ö–æ—Ç—è –±—ã –∫—Ä–∞—Ç–∫–∏–π –æ—Ç–≤–µ—Ç.")

    print("‚úÖ –ë–ª–æ–∫ ¬´–¶–µ–ª–∏¬ª —Å–æ–±—Ä–∞–Ω. –ü–µ—Ä–µ—Ö–æ–¥–∏–º –∫ —Å–ª–µ–¥—É—é—â–µ–º—É —à–∞–≥—É!")
    return answers
```

```python
def generate_resume_and_cover_letter(about_me, goals):
    # –§–æ—Ä–º–∏—Ä—É–µ–º —Ç–µ–∫—Å—Ç —Ä–µ–∑—é–º–µ
    resume_text = f"""
–û —Å–µ–±–µ:
- –§–ò–û: {about_me['q1']}
- –ì–æ—Ä–æ–¥: {about_me['q3']}
- –ö–æ–Ω—Ç–∞–∫—Ç—ã: {about_me['q2']}, {about_me['q4']}
- –£—Ä–æ–≤–µ–Ω—å: {about_me['q5']}
- –°–ø–µ—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è: {about_me['q6']}
- –í –∏–∑—É—á–µ–Ω–∏–∏: {about_me['q7']}
- –•–æ–±–±–∏: {about_me['q8']}

–¶–µ–ª—å:
–ò—â—É –ø–æ–∑–∏—Ü–∏—é {goals['g1']} –≤ –∫–æ–º–ø–∞–Ω–∏–∏, —Ä–∞–±–æ—Ç–∞—é—â–µ–π –≤ —Å—Ñ–µ—Ä–µ {goals['g2']}, —á—Ç–æ–±—ã –ø—Ä–∏–º–µ–Ω—è—Ç—å —Å–≤–æ–∏ –Ω–∞–≤—ã–∫–∏ –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –∏–Ω—Ç–µ—Ä–µ—Å–Ω—ã—Ö –∑–∞–¥–∞—á. –í–∞–∂–Ω—ã {goals['g3']}. –ü—Ä–µ–¥–ø–æ—á—Ç–∏—Ç–µ–ª—å–Ω–æ —Ä–∞–±–æ—Ç–∞—Ç—å –≤ {goals['g4']}.
"""

    print("\nüìÑ –í–∞—à–µ —Ä–µ–∑—é–º–µ (—á–µ—Ä–Ω–æ–≤–∏–∫):")
    print(resume_text)

    # –ó–∞–ø—Ä–æ—Å –≤–∞–∫–∞–Ω—Å–∏–∏
    print("\nüìé –í—Å—Ç–∞–≤—å—Ç–µ —Ç–µ–∫—Å—Ç –≤–∞–∫–∞–Ω—Å–∏–∏, –Ω–∞ –∫–æ—Ç–æ—Ä—É—é –≤—ã –æ—Ç–∫–ª–∏–∫–∞–µ—Ç–µ—Å—å:")
    vacancy = input("–í–∞–∫–∞–Ω—Å–∏—è: ")

    # –ì–µ–Ω–µ—Ä–∞—Ü–∏—è —Å–æ–ø—Ä–æ–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ–≥–æ –ø–∏—Å—å–º–∞ –∏ –æ—Ü–µ–Ω–∫–∏
    prompt = f"""
–ù–∞ –æ—Å–Ω–æ–≤–µ —Å–ª–µ–¥—É—é—â–µ–≥–æ —Ä–µ–∑—é–º–µ –∏ —Ç–µ–∫—Å—Ç–∞ –≤–∞–∫–∞–Ω—Å–∏–∏:

–†–ï–ó–Æ–ú–ï:
{resume_text}

–í–ê–ö–ê–ù–°–ò–Ø:
{vacancy}

–í—ã–ø–æ–ª–Ω–∏ –¥–≤–µ –∑–∞–¥–∞—á–∏:

1. –ù–∞–ø–∏—à–∏ –∫–æ—Ä–æ—Ç–∫–æ–µ —Å–æ–ø—Ä–æ–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ–µ –ø–∏—Å—å–º–æ (–¥–æ 150 —Å–ª–æ–≤) –æ—Ç –ª–∏—Ü–∞ –∫–∞–Ω–¥–∏–¥–∞—Ç–∞. –ü–∏—Å—å–º–æ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –ø–µ—Ä—Å–æ–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–º, —É–≤–∞–∂–∏—Ç–µ–ª—å–Ω—ã–º –∏ –ø–æ–¥—á—ë—Ä–∫–∏–≤–∞—Ç—å —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è–º.

2. –û—Ü–µ–Ω–∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ —Ä–µ–∑—é–º–µ –≤–∞–∫–∞–Ω—Å–∏–∏ –ø–æ —à–∫–∞–ª–µ –æ—Ç 1 –¥–æ 10. –£–∫–∞–∂–∏:
   - –°–∏–ª—å–Ω—ã–µ —Å—Ç–æ—Ä–æ–Ω—ã (—á—Ç–æ —Å–æ–≤–ø–∞–¥–∞–µ—Ç),
   - –ü—Ä–æ–±–µ–ª—ã –∏–ª–∏ –Ω–µ—Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏—è,
   - –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏ –ø–æ —É–ª—É—á—à–µ–Ω–∏—é —Ä–µ–∑—é–º–µ –ø–æ–¥ —ç—Ç—É –≤–∞–∫–∞–Ω—Å–∏—é.
"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.4,
        max_tokens=500
    )

    result = response.choices[0].message.content.strip()
    print("\n" + "="*60)
    print("‚úÖ –†–ï–ó–£–õ–¨–¢–ê–¢:")
    print("="*60)
    print(result)
```

```python
# –°–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö
about_me = collect_about_me()
goals = collect_goals()

# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –∏ –∞–Ω–∞–ª–∏–∑
generate_resume_and_cover_letter(about_me, goals)
```
